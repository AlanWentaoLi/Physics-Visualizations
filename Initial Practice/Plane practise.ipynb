```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python
import numpy as np
```

```python
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 40)

z1 = [[6 - i - j for i in x] for j in y]

z2 = [[zij+10 for zij in zi] for zi in z1]
z3 = [[zij-10 for zij in zi] for zi in z1]

iplot([
    dict(z=z1, showscale=False, type='surface'),
    dict(z=z2, showscale=False, opacity=0.9, type='surface'),
    dict(z=z3, showscale=False, opacity=0.9, type='surface')])
```

```python
def zCoordinates(a, d):
    x = np.linspace(-10, 10, 50)
    y = np.linspace(-10, 10, 50)
    z = [[(d - a[0]*i - a[1]*j)/a[2] for i in x] for j in y]
    return z

def pointIntersect(n1, d1, n2, d2):
    point = np.linalg.solve([n1[1:], n2[1:]], [d1, d2])
    p = [0, point[0], point[1]]
    return p
```

```python
n1 = [2.0, 3.0, -1.0]
d1 = 7.0
z1 = zCoordinates(n1, d1)

n2 = [1.0, 1.0, 1.0]
d2 = 1.0
z2 = zCoordinates(n2, d2)

gradient = np.cross(n1,n2)

point = pointIntersect(n1, d1, n2, d2)

print(gradient, point)

x = [point[0] + gradient[0]* t for t in np.linspace(-5, 5, 50)]
y = [point[1] + gradient[1]* t for t in np.linspace(-5, 5, 50)]
z = [point[2] + gradient[2]* t for t in np.linspace(-5, 5, 50)]

iplot(
    [dict(z=z1,
          colorscale='Reds',
          showscale=False,
          type='surface'),
     dict(z=z2,
          colorscale='Blues',
          showscale=False,
          type='surface'),
     go.Scatter3d(x = x,
                  y = y,
                  z = z,
                  mode = 'lines'
                 ),
     go.Scatter3d(x = 0,
                  y = 2,
                  z = -1,
                  mode='markers',
                  marker = dict(size=50, color ='Blacks'))]
)

```
